## datetime Index

```python
df_tesla_custom=df_tesla.reset_index()
```

```python
df_tesla_custom.head()
```

## set index as date

```python
df_tesla_custom.set_index('Date', drop=True)
```

## resampling

```python
df_tesla.resample(rule='MS').max()['Open'].plot(figsize=(12,4))
```

## rolling

```python
df_tesla['open:30day_rolling']=df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head()
```

```python
df_tesla[['Open','open:30day_rolling']].plot(figsize=(12,4))
```

# Simple Moving Average

```python
df_tesla['Open:10daysRolling']=df_tesla['Open'].rolling(window=10,min_periods=1).mean()
df_tesla['Open:20daysRolling']=df_tesla['Open'].rolling(window=20,min_periods=1).mean()
df_tesla['Open:30daysRolling']=df_tesla['Open'].rolling(window=30,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10daysRolling','Open:20daysRolling','Open:30daysRolling']].plot(figsize=(12,4), xlim=['2024-09-06', '2023-09-06'])
```

# Cumulative Moving Avergae

```python
##cma
df_tesla['Open:CMA']=df_tesla['Open'].expanding().mean()
```

```python
df_tesla[['Open:CMA']].plot()
```

# Exponential weighted Moving Average

```python
df_tesla['EMA_0.1']=df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
df_tesla['EMA_0.3']=df_tesla['Open'].ewm(alpha=0.3, adjust=False).mean()
df_tesla['EMA_0.5']=df_tesla['Open'].ewm(alpha=0.5, adjust=False).mean()
df_tesla['EMA_0.1']=df_tesla['Open'].ewm(alpha=0.1, adjust=False,span=5).mean()
df_tesla['EMA_0.3']=df_tesla['Open'].ewm(alpha=0.3, adjust=False,span=5).mean()
df_tesla['EMA_0.5']=df_tesla['Open'].ewm(alpha=0.5, adjust=False,span=5).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3','EMA_0.5']].plot(figsize=(12,4), xlim=['2024-09-06', '2023-09-06'])
```
